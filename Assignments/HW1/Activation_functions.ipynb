```python
# import libraries
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# create some dummy data
test_data_np = np.linspace(-10.0, 10.0, num=500)
test_data_torch = torch.tensor(test_data_np, dtype=torch.float64)
```

```python
# define a function for plotting
def plot_function(x, y):
    plt.plot(x, y)
```

```python
# define a function to test the two activation functions
def test(x, function1, function2):
    output_fn1 = function1(x)
    output_fn2 = function2(x)
    return torch.allclose(output_fn1,output_fn2), output_fn1
```

# Sigmoid

```python
# implement the sigmoid activation function
##### TO DO #####
def my_sigmoid(inputs):
    """
      given inputs which can be a number or a tensor, calculates sigmoid activation function
      Formula for Sigmoid function is: sigmoid(z) = 1 / (1 + e^(-z))
    """
    return 1 / (1 + np.exp(-inputs))
```

```python
# test your function against the sigmoid activation function from torch
result_sigmoid, output_sigmoid = test(test_data_torch, my_sigmoid, nn.Sigmoid())
print(result_sigmoid)
```

```python
# plot the sigmoid activation function
plot_function(test_data_np, output_sigmoid)
```

# Softmax

```python
# implement the softmax activation function
##### TO DO #####
def my_softmax(inputs):
    """
      given inputs which can be a number or a tensor, calculates softmax activation function
      Formula for Softmax function is: softmax(oi) = e^(oi) / (e^(o1) + e^(o2) + ... + e^(on))
    """
    exp_of_elems = np.exp(inputs)
    sum_of_exp = np.sum(exp_of_elems.numpy())
    return exp_of_elems / sum_of_exp
```

```python
# test your function against the softmax activation function from torch
result_softmax, output_softmax = test(test_data_torch, my_softmax, nn.Softmax())
print(result_softmax)
```

```python
# plot the softmax activation function
plot_function(test_data_np, output_softmax)
```

# ReLU

```python
# implement the ReLU activation function
##### TO DO #####
def my_relu(inputs):
    """
      given inputs which can be a number or a tensor, calculates Relu activation function
      Formula for Relu function is: Relu(z) = Max(0,z)
    """
    inputs[inputs < 0] = 0
    return inputs
```

```python
# test your function against the ReLU activation function from torch
result_relu, output_relu = test(test_data_torch, my_relu, nn.ReLU())
print(result_relu)
```

```python
# plot the ReLU activation function
plot_function(test_data_np, output_relu)
```

# Leaky ReLU

```python
# implement the Leaky ReLU activation function
##### TO DO #####
def my_leaky_relu(inputs):
    """
      given inputs which can be a number or a tensor, calculates Leaky Relu activation function
      Formula for Leaky Relu function is: leaky_relu(z) = max(min(alpha*z, 0), z)
    """
    alpha = 0.01
    inputs[inputs < 0] = alpha * inputs[inputs < 0]
    return inputs
```

```python
# test your function against the Leaky ReLU activation function from torch
result_leaky_relu, output_leaky_relu = test(test_data_torch, my_leaky_relu, nn.LeakyReLU())
print(result_leaky_relu)
```

```python
# plot the Leaky ReLU activation function
plot_function(test_data_np, output_leaky_relu)
```

# Tanh

```python
# implement the tanh activation function
##### TO DO #####
def my_tanh(inputs):
    """
      given inputs which can be a number or a tensor, calculates Tanh activation function
      Formula for Tanh function is: Tanh(z) = (1 - e^(-2*z)) / (1 + e^(-2*z))
    """
    return (1 - np.exp(-2*inputs)) / (1 + np.exp(-2*inputs))
```

```python
# test your function against the tanh activation function from torch
result_tanh, output_tanh = test(test_data_torch, my_tanh, nn.Tanh())
print(result_tanh)
```

```python
# plot the tanh activation function
plot_function(test_data_np, output_tanh)
```
